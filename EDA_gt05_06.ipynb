**EDA:**

EDA hace referencia a el análisis de datos exploratorios por sus siglas en inglés.

Este sirve para conocer las principales características de un conjunto de datos. Nos ayuda a tener un mayor control sobre los datos y así poder observar cual es su comportamiento, también nos ayuda a saber si el análisis de datos utilizado es adecuado.

Tipos de EDA:


*   Univariante no gráfica: en este caso los datos solo tienen una variable, su objetivo es describir datos y encontrar patrones existentes.

*   Gráfico univariante: este análisis no muestra en totalidad los datos, por lo que es necesario utilziar métodos gráficos como; Diagramas de tallo y hojas, histrogramas, diagramas de caja entre otros.

*   Multivariante no gráfica: Estos datos tienen más de una variable, nos muestra la relación que hay entre las variables por medio de métodos estadísticos.

*  Gráfico multivariante: Utilizan gráficos para mostrar la relación entre dos o más conjuntos de datos. El gráfico que más se utiliza es el diagrama de barras.
